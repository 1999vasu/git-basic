```python
a=np.array(range(0,100,5))
b=np.array(range(0,100,5))
plt.figure(0)
plt.title('line plot')
plt.grid('on')
plt.xlabel('year')
plt.xlim(0,100)
plt.ylabel('no of students')
plt.ylim(0,100)
plt.plot(a,b,'r',label="aug")
plt.scatter(a,2*a,c='g',label='june',marker='*')
plt.legend()
plt.show()
```

```python
x = np.linspace(-2*np.pi,2*np.pi,num=100)
plt.figure()
plt.plot(x,np.sin(x),'r')
plt.scatter(x,np.cos(x),c='g')
plt.grid('on')
plt.show()
```
